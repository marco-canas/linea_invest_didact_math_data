#### 1. **An치lisis descriptivo**
   - Pingouin permite calcular estad칤sticas descriptivas de manera sencilla, como medias, desviaciones est치ndar, intervalos de confianza, entre otros.


   - **Ejemplo**: Calcular la media y desviaci칩n est치ndar de las calificaciones pre y post intervenci칩n.


````python
import pandas as pd  # Importar Pandas para leer el CSV y estad칤sticas descriptivas
import pingouin as pg  # Importar Pingouin para an치lisis estad칤stico avanzado
import numpy as np  # Para generar datos ficticios

# Crear datos ficticios y guardarlos en un archivo CSV
np.random.seed(123)  # Para reproducibilidad
data = {
    'id': range(1, 21),  # 20 estudiantes
    'calificacion_pre': np.random.normal(70, 10, 20).round(1),  # Media 70, desviaci칩n 10
    'calificacion_post': np.random.normal(80, 8, 20).round(1)   # Media 80, desviaci칩n 8
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos_g_lideres.csv', index=False)

# Leer el archivo CSV con Pandas
data = pd.read_csv('datos_g_lideres.csv')

# Estad칤sticas descriptivas usando Pandas
desc_pre = data['calificacion_pre'].describe()
desc_post = data['calificacion_post'].describe()

print("Estad칤sticas pre-intervenci칩n:")
print(desc_pre)
print("\nEstad칤sticas post-intervenci칩n:")
print(desc_post)

# Ejemplo de an치lisis avanzado con Pingouin: Prueba t para muestras relacionadas
t_test = pg.ttest(data['calificacion_pre'], data['calificacion_post'], paired=True)
print("\nPrueba t para muestras relacionadas (pre vs post):")
print(t_test)
````
