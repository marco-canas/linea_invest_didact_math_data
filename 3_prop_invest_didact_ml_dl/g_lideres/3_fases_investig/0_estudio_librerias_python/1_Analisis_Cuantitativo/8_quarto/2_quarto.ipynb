# Qué es quarto  



**Quarto** es una herramienta moderna y poderosa para la creación de documentos dinámicos y reproducibles. Combina texto, código, gráficos y resultados en un solo documento, lo que lo hace ideal para la investigación educativa, especialmente en proyectos como **G-Líderes**, donde la triangulación de datos es fundamental.

---

### ¿Qué es Quarto?

Quarto es un sistema de publicación de código abierto que permite crear documentos, informes, presentaciones y sitios web interactivos. Es compatible con varios lenguajes de programación, como **R**, **Python**, **Julia** y **JavaScript**, y genera salidas en formatos como HTML, PDF, Word, PowerPoint y más.

Algunas características clave de Quarto:
1. **Integración de código y texto**: Permite combinar narrativa con código ejecutable y sus resultados.
2. **Reproducibilidad**: Los documentos son completamente reproducibles, lo que garantiza transparencia en la investigación.
3. **Flexibilidad**: Soporta múltiples lenguajes y formatos de salida.
4. **Colaboración**: Facilita el trabajo en equipo al integrarse con sistemas de control de versiones como Git.
